**Some Questions to solve**
- How does the frequency of mental health illness and attitudes towards mental health vary by geographic location?
- What are the strongest predictors of mental health illness or certain attitudes towards mental health in the workplace?

```python
df = df[df['Country'].notna() & df['treatment'].notna()]

# Calculate treatment frequency by country
treatment_by_country = df.groupby('Country')['treatment'].value_counts(normalize=True).unstack().fillna(0)
treatment_by_country = treatment_by_country.sort_values('Yes', ascending=False).head(10)

# Calculate attitudes by country (mental_health_consequence)
attitudes_by_country = df.groupby('Country')['mental_health_consequence'].value_counts(normalize=True).unstack().fillna(0)
attitudes_by_country = attitudes_by_country.loc[treatment_by_country.index]

# Plotting
sns.set(style="whitegrid")
fig, axes = plt.subplots(2, 1, figsize=(14, 12))

# Plot 1: Treatment frequency
treatment_by_country['Yes'].plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Proportion Receiving Mental Health Treatment by Country (Top 10)')
axes[0].set_ylabel('Proportion')
axes[0].set_xlabel('Country')
axes[0].tick_params(axis='x', rotation=45)

# Plot 2: Mental health consequence perceptions
attitudes_by_country.plot(kind='bar', stacked=True, ax=axes[1], colormap='Set2')
axes[1].set_title('Perceived Mental Health Consequences in the Workplace by Country (Top 10)')
axes[1].set_ylabel('Proportion')
axes[1].set_xlabel('Country')
axes[1].legend(title='Consequence Perceived')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Select relevant features
features = [
    'Age', 'Gender', 'self_employed', 'family_history', 'work_interfere',
    'no_employees', 'remote_work', 'tech_company', 'benefits',
    'care_options', 'wellness_program', 'seek_help', 'anonymity'
]
```

```python
# Drop rows with missing values in those columns
df_model = df[features + ['treatment']].dropna()
```

```python
# Encode categorical variables
df_encoded = pd.get_dummies(df_model.drop('treatment', axis=1), drop_first=True)
```

```python
# Encode target
df_encoded['treatment'] = df_model['treatment'].map({'Yes': 1, 'No': 0})
```

```python
# Split features and target
X = df_encoded.drop('treatment', axis=1)
y = df_encoded['treatment']
```

```python
# Fit Random Forest
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Feature importance plot
importances = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=importances.values, y=importances.index)
plt.title('Feature Importance for Predicting Mental Health Treatment')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

**insights:** The feature importance analysis using a Random Forest model reveals that the strongest predictors of seeking mental health treatment in the workplace are having a **family history of mental illness** and experiencing **mental health issues that interfere with work**, indicating personal and functional impacts as key drivers. Additionally, workplace-related factors such as **awareness of care options**, **availability of mental health benefits**, **presence of wellness programs**, and **employer encouragement to seek help** significantly influence attitudes toward treatment. The perception of **anonymity protection** also plays a crucial role, as employees are more likely to seek help when they feel their privacy is safeguarded. Overall, both personal history and organizational support structures are critical in shaping employees' willingness to address mental health concerns.
